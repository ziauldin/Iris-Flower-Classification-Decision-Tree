```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Loading the iris dataset
iris = pd.read_csv("Task 6.csv")

# Display the first few rows of the dataset to confirm structure
print(iris.head(5))
```

```python

# Encoding the target variable
label_encoder = LabelEncoder()
iris['target'] = label_encoder.fit_transform(iris['Species'])

# Dropping the 'Id' and 'Species' columns
X = iris.drop(columns=['target', 'Species', 'Id'])
y = iris['target']

# Verifying the column names of features
print(X.columns)
```

```python
# Defining the decision tree algorithm
dtree = DecisionTreeClassifier()
dtree.fit(X, y)

print('Decision Tree Classifier Created')
```

```python
# Visualizing the decision tree
plt.figure(figsize=(20,10))
# Visualizing the decision tree
plt.figure(figsize=(20,10))
plot_tree(dtree, feature_names=X.columns.tolist(), class_names=label_encoder.classes_.tolist(), filled=True, rounded=True, fontsize=12)
plt.show()
```
